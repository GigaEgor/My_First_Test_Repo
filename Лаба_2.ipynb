```
#  Подключение нужных для работы моделей
import pandas as pd     # для чтения и записи табличных файлов
import numpy as np      # для математической обработки данных
import matplotlib.pyplot as plt # для построения графиков
```

```
#  Получение и чтение файла из облака

url = "https://drive.google.com/uc?export=download&id=18VeFLUIj22hRmc_tNGjfoawyn7iyqQ_w"


input_df = pd.read_csv(url, header=0) 

# вывод заголовка таблицы и первых пяти строк (сэмплов)
input_df.head()
```

```
input_matrix = input_df.values # перевод таблицы в матрицу
```

```
fig, ax = plt.subplots(9, 1, figsize = (15, 12))

timestamps = range(0, input_matrix.shape[0] * 5, 5)

# зададим диапазон
begin = 2
end   = 146 
step = 2

ax[0].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 2], 'black', linestyle = '--', marker = 'o')
ax[1].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 3], '#7FFF00', linestyle = '--', marker = 'o')
ax[2].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 4], '#FA8072', linestyle = '--', marker = 'o')
ax[3].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 5], '#00BFFF', linestyle = '--', marker = 'o')
ax[4].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 6], '#191970', linestyle = '--', marker = 'o')
ax[5].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 7], '#00CED1', linestyle = '--', marker = 'o')
ax[6].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 8], '#008000', linestyle = '--', marker = 'o')
ax[7].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 9], '#2F4F4F', linestyle = '--', marker = 'o')
ax[8].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 10], '#FFD700', linestyle = '--', marker = 'o')


legend = ['coal','nuclear', 'ccgt','wind','pumped','hydro','biomass','oil','solar']


for _ in zip(ax, legend):
  _[0].set_ylabel('Мощность, ГВт')
  _[0].legend([_[1]])
  _[0].grid(True)


ax[-1].set_xlabel('Время, мин')

plt.show()
```

```
generation_sum = (input_matrix[: , 3 : -1]).sum(axis = 0)

labels = input_df.columns[3 : -1]

fig1, ax1 = plt.subplots(figsize = (5, 5))
ax1.pie(generation_sum, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')

plt.show()
```
